## Getting starting

Let's import the libraries required for Python to run Seaborn.

*   *Numpy* is a library for working with arrays and matrices in Python
*   *Pandas* is a library for data analysis and manipulation. Pandas makes it easy to manipulate data files.
*   *Matplotlib* is a library for visualizing data. This library is the basis for Seaborn, but Seaborn is more convenient.
*    *[Seaborn](https://seaborn.pydata.org/)* provides high-level commands for plots and data exploration based on the Matplotlib packages. Although much of the functionality is the same as the  raw Matplotlib commands,  Seaborn is much more convenient.

## Univariate Analysis

### Qualitative Analysis: Count Plot



The most common chart for univariate analysis of categorical data is a bar chart.

```python
sns.set_style('darkgrid')
count_plot = sns.countplot(data=tips, x="day", color="Yellow")
count_plot.set(xlabel="Day", ylabel = "Frequency")
plt.show()
```

#### Color
It is possible to change the color of the chart with the color argument.

#### Sorting bars

In most frequency plots or other bar charts, sorting the categorical axis enables a better analysis of the bars. The order parameter allows you to sort the bars by the columns.

```python
sns.set_style('darkgrid')
count_plot = sns.countplot(data=tips, x="day", color="Yellow", order=tips["day"].value_counts().index)
count_plot.set(xlabel="Day", ylabel = "Frequency")
plt.savefig("count_plot_transparent.png", transparent = True)
plt.show()
```

```python
sns.set_style('darkgrid')
count_plot = sns.countplot(data=tips, x="day", color="Yellow", order=tips["day"].value_counts().index)
count_plot.set(xlabel="Day", ylabel = "Frequency")
plt.savefig("count_plot_highRes.png", dpi=300)
plt.show()
```

### Quantitative Analysis: Boxplot

Boxplot charts enable the analysis of the variable distribution according to the percentiles.

```python
ax = sns.boxplot(x = tips["total_bill"] )
```

Calculating Lower Quantile Total Bill

```python
Lower_quamtile_total_bill = np.quantile(tips["total_bill"], 0.25)
```

Calculating Upper Quantile Total Bill

```python
Upper_quamtile_total_bill = np.quantile(tips["total_bill"], 0.75)
```

Inter-Quantile Range for the total_bill

```python
Iqr = Upper_quamtile_total_bill - Lower_quamtile_total_bill
```

Checking the Inter-Quantile Range for the total_bill

```python
Iqr
```

```python
# Checking the upper fence total bill
total_upper_fence_bill = np.min([np.max(tips["total_bill"]), Upper_quamtile_total_bill + 1.5 * Iqr] )
print(total_upper_fence_bill)
```

```python
# Checking the lower fence total bill
total_lower_fence_bill = np.max([np.min(tips['total_bill']), Lower_quamtile_total_bill - 1.5 * Iqr])
total_lower_fence_bill
```

### Quantitative Analysis: Strip plot

This plot shows the points on the x-axis and a random point on the y-axis within a specificed range (the jitter parameter). These plots are typically used with a categorical variable on the other axis, but they can be used alone as well.

```python
tips = sns.load_dataset('tips')
strip_plot_val = sns.stripplot(data = tips, x='total_bill', jitter=0.05)
```

# **Strip-Plot :- Create a strip plot, a box plot, and a point plot showing the distribution of the carats variable in the diamonds data set.**

```python
# Importing the data of Diamonds Dataset
Diamond_data = sns.load_dataset('diamonds')
Diamond_data.head()
```

```python
# Strip-Plot plotting using matplot-lib
# ploting_data.text(xlabel='Carat Value', ylabel='Distribution Value', s='The Carat value Distribution Using Strip-Plot')




import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Assuming Diamond_data is your dataset
# Strip plot
strip_plot = sns.stripplot(data=Diamond_data, x='carat', jitter=0.1, alpha=0.01, color='blue')

# Box plot (use the same axis as the strip plot)
sns.boxplot(data=Diamond_data, x='carat', color='blue', ax=plt.gca(), width=0.3)

# Setting the Alignment of the text and designing the Interface
strip_plot.text(

                x = 0.5,
                y = 1.1,
                s = 'Diamonds Over The 2 Carats Are The Outliers',
                va = 'bottom',
                ha = 'center',
                fontsize = 12,
                transform = strip_plot.transAxes


)

# Setting the Alignment of the text and designing the Interface
strip_plot.text(

                x = 0.5,
                y = 1.05,
                s = 'Source : Diamond Dataset ( Sea-Born )',
                va = 'bottom',
                ha = 'center',
                fontsize = 8,
                transform = strip_plot.transAxes


)


# Point plot (use the same axis as well)
sns.pointplot(data=Diamond_data, x='carat', estimator=np.mean, color='red', markers="X", ax=plt.gca())

# Adjust labels and limits
plt.xlabel('Carat Value')
plt.xlim(0, 6)

# Show the plot
plt.show()
```
