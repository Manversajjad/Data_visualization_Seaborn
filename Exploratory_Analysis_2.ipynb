```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head(n=4)
```

```python
tips_pivoted = pd.pivot_table(data=tips, index=['day'], columns=['time'],
                              values='total_bill', aggfunc=len)
```

```python
print(tips_pivoted)
```

```python
ax = sns.heatmap(data=tips_pivoted)
```

```python
ax = sns.heatmap(data=tips_pivoted, cmap="Greens")
sns.heatmap()
```

```python
tips_pivoted_filled = pd.pivot_table(tips, values="total_bill",
                                     index=['day'], columns=['time'],
                                     aggfunc=len,
                                     fill_value=0)
```

```python

ax = sns.heatmap(data=tips_pivoted_filled)
```

```python
ax = sns.heatmap(tips_pivoted_filled, cmap="Greens", annot=True)
```

```python
ax = sns.scatterplot(data=tips, x="total_bill", y="tip", color="green")
```

```python
ax = sns.regplot(data=tips, x="total_bill", y="tip", color="green")
```

### Correlations

```python
corr = diamonds.corr()
print(corr)
```

```python
sns.heatmap(data=corr)
```

```python
Ddiv_palette = sns.diverging_palette(230,20,as_cmap=True)
sns.heatmap(data=corr, cmap=div_palette, linewidths=1, center=0, vmax=1, vmin=-1)
plt.show()
```

```python
mask1 = np.triu(np.ones_like(corr, dtype=bool))
div_palette = sns.diverging_palette(230,20,as_cmap=True)
sns.heatmap(data=corr, mask = mask1, cmap=div_palette, linewidths=1, center=0, vmax=1, vmin=-1)
plt.show()
```
