```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mkt = pd.read_csv(r'/content/supermarket_sales new.csv')
mkt.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Strip plot
mkt_d = sns.stripplot(data=mkt, x='Unit price', alpha=0.1, jitter=0.1, color='blue')

# Box plot (use the same axis as the strip plot)
sns.boxplot(data=mkt, x='Unit price', color='blue', ax=plt.gca(), width=0.3)

# Point_plot
mkt_p = sns.pointplot(data=mkt, x='Unit price', markers='X', color='red')

# Adjust labels and limits
plt.xlim(0, 120)           # Adjust x-axis limit

# Setting the Alignment of the text and designing the Interface
mkt_d.text(

                x = 0.5,
                y = 1.1,
                s = 'Marketing Sales Data Defined By Price',
                va = 'bottom',
                ha = 'center',
                fontsize = 12,
                transform = mkt_d.transAxes


)

# Setting the Alignment of the text and designing the Interface
mkt_d.text(

                x = 0.5,
                y = 1.05,
                s = 'Source : Marketing Sales Dataset ( Kaggle )',
                va = 'bottom',
                ha = 'center',
                fontsize = 8,
                transform = mkt_d.transAxes


)

# Show the plot
plt.show()

```

```python
density_plot_dia = sns.kdeplot(data=mkt, x='Unit price', color='grey', cumulative=True)
density_plot_dia.set(ylabel='price', xlabel = 'Price Measure' )
```
